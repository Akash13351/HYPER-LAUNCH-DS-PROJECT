# **MODEL DEVELOPMENT**

```python
df_encoded = pd.get_dummies(df, drop_first=True)
x = df_encoded.drop(columns=['Calories_Burned'])
y = df_encoded['Calories_Burned']
```

IN THIS X VARIABLE CONTAINS THE INPUT FEATURES BY DROPPING THE TARGET VALUES.Y VARIABLE CONTAINS ONLY THE TARGET VALUE WHISCH IS "CALORIES BURNED"

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

  Here  train_test_split() is a function from sklearn that splits the dataset into the following:


1.  **x_train:** It contains 80 % of the input data to train the model.
2.   **x_test:** It contains 20 % of the input data to test the model.
3. **y_train:** It contains 80 % of the target variable.
4.   **y_test:** It contains 20 % of the target variable.



```python
from sklearn.linear_model import LinearRegression
```

HERE,CALORIES BURNED IS THE CONTINUOUS NUMERIC VALUE SO WE USE LINEAR REGRESSION
FOR THIS DATAFRAME

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

IT INITIALIZE AND TRAIN THE MODEL

```python
y_pred=model.predict(x_test)
```

THIS PROCESS PREDICTS THE TEST DATA

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score
import numpy as np

mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
r2=r2_score(y_test,y_pred)
```

#**RESULT PRESENTATION**

SHOW THE MEAN ABSOLUTE ERROR,MEAN SQUARED ERROR,ROOT MEAN SQUARED ERROR,R-SQUARED

```python
print("Mean Absolute Error:",mae)
print("Mean Squared Error:",mse)
print("Root Mean Squared Error:",rmse)
print("R-squared:",r2)
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test,y_test,color='RED')
plt.scatter(y_test,y_pred,color='VIOLET')
plt.plot([y.min(),y.max()],[y.min(),y.max()],'--b',label='perfect prediction line')
plt.xlabel("Actual Calories_Burned")
plt.ylabel("Predicted Calories_Burned")
plt.title("Actual vs Predicted Calories_Burned")
plt.show()
```

IT SHOWS THAT HOW WELL THE MODEL PREDICTED THE CALORIES BURNED BY THE MEMBERS AS THE ACTUAL CALORIES BURNED BY THE MEMBERS.
